```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
```

```python
data=pd.read_csv("iris.csv")
```

```python
data.info()
```

```python
data
```

```python
data.set_index("Id",inplace=True)
```

```python
data
```

```python
data.isnull().sum()
```

```python
y_train = data['Species']
y_train
```

```python
X_train=data.drop("Species",axis=1)
print(X_train)
```

```python
X_train_sub,X_test_sub,y_train_sub,y_test_sub=train_test_split(X_train,y_train,random_state=0)
```

```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier
base_classifier = DecisionTreeClassifier(max_depth=1)
adaboost_classifier = AdaBoostClassifier(base_classifier, n_estimators=20, random_state=42)
adaboost_classifier.fit(X_train_sub, y_train_sub)
print("Accuracy score (training): {:.3f}".format(adaboost_classifier.score(X_train_sub, y_train_sub)))
print("Accuracy score (testing): {:.3f}".format(adaboost_classifier.score(X_test_sub, y_test_sub)))
```
